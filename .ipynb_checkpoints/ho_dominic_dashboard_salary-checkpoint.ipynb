# Preprocessing

```python
dataset = pd.read_csv('Salary.csv')
dataset.head(100)
```

```python
# cache data to improve dashboard performance
if 'data' not in pn.state.cache.keys():

    dataset = pd.read_csv('Salary.csv')

    pn.state.cache['data'] = dataset.copy()

else: 

    dataset = pn.state.cache['data']
```

```python
#one hot encoding 'Sex'

dataset.replace({'Sex': {'M': 1, 'F': 0}},inplace=True)
#dataset.replace({'Department': {'IT/IS': 1, 'Software Engineering': 2,'Production': 3, 'Sales': 4,'Admin Offices':5,'Executive Office':6  }},inplace=True)

dataset.drop(['EmpStatusID'],axis=1, inplace=True)
dataset.drop(['DeptID'],axis=1, inplace=True)
dataset.drop(['FromDiversityJobFairID'],axis=1, inplace=True)



```

```python
dataset.head(20)
```

```python

# Make DataFrame Pipeline Interactive
idf = dataset.interactive()
```

# Salary/engagement by age by department

```python
# Define Panel widgets
age_slider = pn.widgets.IntSlider(name='Age slider', start=0, end=100, step=1, value=30)
age_slider
```

```python
# Radio buttons for Salary measures
yaxis_salary = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['Salary', 'EngagementSurvey',],
    button_type='success'
)
```

```python
continents = dataset['Salary']

salary_pipeline = (
    idf[
        (idf.Age <= age_slider) 
        
    ]
    .groupby(['Age','Department'])[yaxis_salary].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Age')  
    .reset_index(drop=True)
)
```

```python
salary_pipeline
```

```python
salary_plot = salary_pipeline.hvplot(x = 'Age', by='Department', y=yaxis_salary,line_width=2, title="Salary by Age by department")
salary_plot
```

# 2) Table -  Salary/Engagement by Age by department

```python
salary_table = salary_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
salary_table
```

# 3) Bar Chart -  Salary/Engagement/Sex(gender) by Age

```python
yaxis_salary_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['Salary', 'EngagementSurvey','Sex'], 
    button_type='success'
)

continents_excl_world = dataset['Department']

salary_source_bar_pipeline = (
    idf[
        (idf.Age == age_slider) &
        (idf.Department.isin(continents_excl_world))
    ]
    .groupby(['Age', ])[yaxis_salary_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='Age')  
    .reset_index(drop=True)
)
```

```python
salary_source_bar_plot = salary_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='Age', 
                                                     y=yaxis_salary_source, 
                                                     title='Salary/EngagementSurvey/Sex(gender) by age')
salary_source_bar_plot
```
